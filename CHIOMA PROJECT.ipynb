```python
# dropping the Unnamed columns
col_to_drop = ['Unnamed: 0']
df = df.drop(columns=col_to_drop, axis=0)
```

```python
# dropping mp_commoditysource column because of the null values 
col_to_drop = ['mp_commoditysource']
df = df.drop(columns=col_to_drop, axis=-1)
```

```python
df.shape
```

```python
#to find the null values

df.isnull().sum()
```

```python
# drop rows null values in state
df = df.dropna(subset=['state'])
```

```python
df.isnull().sum()
```

```python
#datatype

df.dtypes
```

```python
# columns

df.columns
```

```python
cols_to_drop = ['country_id', 'state_id','market_id','produce_id','currency_id','pt_id','um_unit_id']

df = df.drop(columns=cols_to_drop, axis=1)
```

```python
df
```

```python
df.shape
```

```python
print(df['produce'][0])
```

```python
#split on the left most parenthesis as the delimeter


temp = df['produce'].str.split("-", n=1, expand =True)
```

```python
temp
```

```python
#replace the produce column in df with the temp[0]


df['produce']= temp[0]
```

```python
df.head()
```

```python
df.tail()
```

```python
df['quantity'] = df['quantity'].replace(['250 G', 'G', '100 ML', 'ML'])

print(df)
```

```python
df.dtypes
```

```python
#Reenaming the Columns
df = df.rename(columns= { 'country' : 'Country' , 'state': 'State', 'market' : 'Market', 
                         'produce': 'Produce', 'currency': 'Currency',
                         'market_type':'Market_Type',
                         'quantity':'Quantity', 'month':'Month', 
                         'year':'Year', 'price':'Price'})
```
